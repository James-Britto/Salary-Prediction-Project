## <center>Feature Encoding and Transformation

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder # importing neccesary liberaries for encoding and transformation
from sklearn.preprocessing import StandardScaler

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from category_encoders import BinaryEncoder

from sklearn.model_selection import train_test_split,cross_val_score
```

```python
X=df.drop('Salary',axis=1) #splitting the data for training and testing the model
y=df['Salary']
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
num_cols=['Years_of_Experience']  # creating pipelines for feature encoding and numeric feature standardization
cat_cols=['Gender','Job_Title']
ord_cols=['Education_Level']

num_pipe=Pipeline(steps=[('standard',StandardScaler())],)
ord_pipe=Pipeline(steps=[('ordinal',OrdinalEncoder(categories=[['High School','Bachelor', 'Masters', 'PhD']]))],)
cat_pipe=Pipeline(steps=[('cat',BinaryEncoder(cols=cat_cols))],)

preprocessor=ColumnTransformer(transformers=[('numeric',num_pipe,num_cols),('ordin',ord_pipe,ord_cols),('categ',cat_pipe,cat_cols)],remainder='passthrough')
```

```python
X_train=preprocessor.fit_transform(X_train) #transforming the train and test data using pipeline
X_test=preprocessor.transform(X_test)
```

# <center>Model Selection

```python
from sklearn.linear_model import Lasso #importing the models from sklearn package

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_squared_error
```

# <center>Performance Evaluation

```python
models={'Lasso Regression':ls,'Random Forest Regressor':rfc,'XGBoost Regressor':xgb,'SVM Regressor':svr,'KNeighbors Regressor':knn}
train_performance={}
test_performance={}

for name,model in models.items():
    y_pred_train=model.predict(X_train)
    y_pred_test=model.predict(X_test)
    print(f'\033[1m\033[94m{name} :\033[0m')
    print('-'*100)
    print('\033[1mTraining Data :\033[0m\t\t\t\t\t\033[1mTest Data :\033[0m')
    print('='*100)
    print(f'RMSE: {round(np.sqrt(mean_squared_error(Y_train,y_pred_train)),2)}\t\t\t\t\tRMSE: {round(np.sqrt(mean_squared_error(Y_test,y_pred_test)),2)}')
    print(f'R2 Score: {round(r2_score(Y_train,y_pred_train),2)}\t\t\t\t\tR2 Score: {round(r2_score(Y_test,y_pred_test),2)}')
    print(f'MAPE: {round(mean_absolute_percentage_error(Y_train,y_pred_train),2)}\t\t\t\t\tMAPE: {round(mean_absolute_percentage_error(Y_test,y_pred_test),2)}')
    print(f'Cross Validation Score : {round(cross_val_score(model,X_train,Y_train).mean(),2)}\t\t\tCross Validation Score : {round(cross_val_score(model,X_test,Y_test).mean(),2)}')
    print()
```
